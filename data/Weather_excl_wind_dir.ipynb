# Scraping weather data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import regex as re
```

### Function to get the weather

```python
def weather(urls, dates):
    """
    input: 
    urls = list of urls (elements as strings)
    dates = list of dates (elements as strings)
    
    output:
    a tuple with five lists of weather data: temp (average daily temp), rain_mm, rain_hrs, wind_spd
    
    
    """
    temp_list = []
    rain_mm_list = []
    rain_hrs_list = []
    wind_spd_list = []
    
    for i in url_list:
        # webscraping
        url = i
        html = requests.get(url)
        content = requests.get(url).content
        soup = BeautifulSoup(content, 'html')

        # find the right row in the temp table    
        table_temp = soup.find_all('table')[0]
        rows = table_temp.find_all('tr')
        row = rows[3]

        #get the right temp
        temp_row = row.text.strip().split('\n')
        avg_temp = temp_row[1]

        #append temp to avg_temp list
        temp_list.append(avg_temp)

        # find the right row in the rain table
        table_rain = soup.find_all('table')[2]
        rain_rows = table_rain.find_all('tr')
        rain_row = rain_rows[2]

        #get the right values
        rain_row = rain_row.text.strip().split("\n")
        rain_mm = rain_row[1]
        rain_hrs = rain_row[3]

        #append rain_mm and rain_hours to their lists
        rain_mm_list.append(rain_mm)
        rain_hrs_list.append(rain_hrs)

        # find the right row in the rain table
        table_wind = soup.find_all('table')[1]
        wind_rows = table_wind.find_all('tr')
        wind_row = wind_rows[2]

        #get the right values
        wind_row = wind_row.text.strip().split("\n")
        wind_spd = wind_row[0]
        wind_spd_list.append(wind_spd)
    return(temp_list, rain_mm_list, rain_hrs_list, wind_spd_list)
```

### Combine the dataframes to one big dataframe

```python
# Put the data in one big dataframe
list_of_dfs = [weather_DeBilt, weather_Twenthe, weather_Eelde, weather_Lauwersoog, weather_Lelystad, weather_DH, weather_HvH, weather_Vlissingen, weather_Ell, weather_Arcen]
weather_df = pd.concat(list_of_dfs)

#display all rows
pd.set_option('display.max_rows', weather_df.shape[0]+1)

#check it out!
weather_df
```

```python
# remove the units
weather_df["rain_mm"] = weather_df["rain_mm"].apply(lambda x: x[0:-3])
weather_df["rain_hrs"] = weather_df["rain_hrs"].apply(lambda x: x[0:-4])
weather_df["wind_spd"] = weather_df["wind_spd"].apply(lambda x: x[0:-4])

weather_df.head(15)
```

```python
# change the strings in columns 1 ,2, 3 and 4 tp int and/or floats
# rain_mm has <0.05 values. These can't be changed into int's or floats. 
# change those to 0.025 first
weather_df.loc[weather_df.rain_mm == "<0,05", "rain_mm"] = "0.025"

weather_df[["temp", "rain_mm", "rain_hrs", "wind_spd"]] = weather_df[["temp", "rain_mm", "rain_hrs", "wind_spd"]].apply(pd.to_numeric)

#change the date column into a datetime object
weather_df["date"]=pd.to_datetime(weather_df["date"])
```
