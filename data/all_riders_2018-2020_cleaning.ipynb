# RATN: cleaning 2018, 2019 and 2020

The webscraping and some basic cleaning is done in the jupyter notebooks in the data-folder. They result in 5 csv-files:
the leaderboards for 2018, 2019, 2020 and 2021 each have a csv, and so has the age of the 2021 riders.

The issue with the leaderboards are the Timings. These are in a horrible format. 
The format for 2018 till 2020 are the same, the format for 2021 is different. 

In this notebook the csv's with the leaderboards 2018 till 2020 are concatenated and the Timings are put into a datetime-format.

```python
# The timings-columns are unworkable. These have to be onverted to the correct datetime-format (YYYY-MM-DD HH:MM)
# Write a function that can take the values of 2 columns and creates a daytime object with the correct date and time

def date(day, time, ref_date):
    """
    input: 
    - day as string: Sun / Mon / Tue / Wed / Thu / Fri / Sat
    - time as string: hh:mm
    - ref_date: as string - the reference date according to which you calculate the outcome
    
    output: 
    - if none of the input-vars is NaN: the date in a datetime-format (YYYY-MM-DD HH:MM)
    - if one the input-vars is NaN: "customize me" as string
    """
    # ---------- MAKE THE DATE -----------
    #make a datetime object of the start_date
    
    #put "empty" in the cells that are empty for a good reason
    if ref_date == 'nan': 
        s = 'empty'
        return s
    if time == 'nan':
        s = "empty"
        return s
    if day == 'nan':
        s = "empty"
        return s
    
    #mark the cells that should have a value with 'customize me'
    if ref_date == "empty": 
        s = 'customize me'
        return s
    if time == "empty":
        s = "customize me"
        return s
    if day == "empty":
        s = "customize me"
        return s
    
    #when there is a day and time and a reference date:
    else:
        #if type of ref_date is datetime object
        if type(ref_date) != str:
            ref_date = str(ref_date)
        else:
            ref_date = ref_date
        #if type of ref_date is string
        d = datetime.strptime(ref_date, '%Y-%m-%d %X')

        #split the datetime object (d), it will now become a list
        splt = ref_date.split('-')

        # week number based on start_date (of d). Note: sunday as the first day of the week
        week_nr = d.strftime("%U")

        # weekday based on start_date (of d). Note: sunday as first day of the week
        day_nr = d.strftime("%w")

        # getting the weekday list --> ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
        # import calendar
        day_list = list(calendar.day_abbr)
        day_list.insert(0, day_list.pop())

        # 6 is end of the week, so new week nr should be 1 higher
        if day_nr == '6' and day != 'Sat':
            new_week_nr = int(week_nr) + 1
        else:
            new_week_nr = int(week_nr)

        new_day = day  # the new recorded day, (inputted into the function by you)
        day_index = day_list.index(day) # get index position of the new_day in day_list

        # take the first element (index 0) of the list, add -W get the new week number add - and the stringday-number 
        new_date = f'{splt[0]}-W{new_week_nr}-{str(day_index)}'

        #create the new datetime object (r):
        #r = datetime.strptime(new_date, "%Y-W%U-%w")

        #------------ ADD THE TIME -------------
        new_date_time = f'{splt[0]}-W{new_week_nr}-{str(day_index)}{str(time)}' 
        s = datetime.strptime(new_date_time, "%Y-W%U-%w%H:%M")
        return s
```

```python
# The timings-columns are unworkable. These have to be onverted to the correct datetime-format (YYYY-MM-DD HH:MM)
# Write a function that can take the values of 2 columns and creates a daytime object with the correct date and time

def customize(day, time, prev_ref_date, DateTime):
    """
    input: 
    - day as string: Sun / Mon / Tue / Wed / Thu / Fri / Sat
    - time as string: hh:mm
    - prev_ref_date: as string - the reference date according to which you calculate the outcome.
    
    output: 
    - if the original DateTime said "customize me" it will give a date in a datetime-format (YYYY-MM-DD HH:MM)
    - if the prev_ref_date is also "empty" it will return "check me out"
    - otherwise it will return the original DateTime 
    """
    # ---------- MAKE THE DATE -----------
    #only apply the function if DateTime == "customize me", otherwise, let the value alone.
    if DateTime == "customize me":
        if prev_ref_date != "empty":   
        # prev_ref_date can either be a date, or empty.
        # if empty, keep empty but indicate that you've changed something
            if DateTime == "customize me":
                #if type of ref_date is datetime object, make it into a string
                if type(prev_ref_date) != str:
                    prev_ref_date = str(prev_ref_date)
                else:
                    prev_ref_date = prev_ref_date

                #now type of ref_date is string
                d = datetime.strptime(prev_ref_date, '%Y-%m-%d %X')

                #split the datetime object (d), it will now become a list
                splt = prev_ref_date.split('-')

                # week number based on start_date (of d). Note: sunday as the first day of the week
                week_nr = d.strftime("%U")

                # weekday based on start_date (of d). Note: sunday as first day of the week
                day_nr = d.strftime("%w")

                # getting the weekday list --> ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
                # import calendar
                day_list = list(calendar.day_abbr)
                day_list.insert(0, day_list.pop())

                # 6 is end of the week, so new week nr should be 1 higher
                if day_nr == '6' and day != 'Sat':
                    new_week_nr = int(week_nr) + 1
                else:
                    new_week_nr = int(week_nr)

                new_day = day  # the new recorded day, (inputted into the function by you)
                day_index = day_list.index(day) # get index position of the new_day in day_list

                # take the first element (index 0) of the list, add -W get the new week number add - and the stringday-number 
                new_date = f'{splt[0]}-W{new_week_nr}-{str(day_index)}'

                #create the new datetime object (r):
                #r = datetime.strptime(new_date, "%Y-W%U-%w")

                #------------ ADD THE TIME -------------
                new_date_time = f'{splt[0]}-W{new_week_nr}-{str(day_index)}{str(time)}' 
                s = datetime.strptime(new_date_time, "%Y-W%U-%w%H:%M")
                return s
            else:
                s = "empty2"
                return s
        else:
            s = "you monster!"
            return s
        
    else:
        s = DateTime
        return s
```

```python
# Oh no, monsters show up at the finish line! 
# Monsters are riders that have a day&time in the TimingFinish column, but didn't finish (Place == NaN) 
# Some monsters did finish (but had 2 consecutive missing values). Here Place != NaN.
# put "customize me" and change these by hand (there will be only a few)  

def cheat(monsters, place):
    """
    input: monsters: the finish-column that contains monsters 
        place: the column that contains the Place of the rider (type = float)
    output:
        True monsters are turned to "empty" 
        False monsters are turned to "customize me"
        non-monsters keep their original values
    """
    if monsters == "you monster!" and math.isnan(place) == True:
        s = "empty"
        return s
    if monsters == "you monster!" and math.isnan(place) == False:
        s = "customize me"
        return s
    else:
        s = monsters
        return s
```
