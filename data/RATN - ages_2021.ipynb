#### Get the name & ages from the riders of 2021

Rider info is placed on "cards" in the website that are encoded in javascript. You don't have to click them to see them, but still BS4 cannot grab the information at once. 

```python
url = 'https://www.adventurebikeracing.com/ratn/'

#set that you want to run the chrome headless (behind the scenes, in the background)
chrome_options = Options()
chrome_options.add_argument("--headless")

#insert absolute path of chrome driver
driver = webdriver.Chrome('C:/Users/jetsa/chromedriver.exe', options = chrome_options)
driver.get(url)

#you can use this as input for BS4:
source = driver.page_source

soup = BeautifulSoup(source, 'html')
print(soup.prettify())
```

#### Extract the right information and clean the data.
There is a whole bunch of information, but I'm only interested in the riders name and age. Therefor I only search for these two classes.

```python
membercard = soup.find_all('div', attrs={'class':["um-member-name", "um-member-tagline um-member-tagline-birth_date"]})
membercard
```

```python
messy_rider_info = []

for i in membercard:
    i=i.text.split(",")
    messy_rider_info.append(i)

print (messy_rider_info)
```

```python
print(type(messy_rider_info))
```

```python
cleaner_rider_info = []

for i in messy_rider_info:
    for j in i:
        cleaner_rider_info.append(re.sub("\t", "", j))
cleaner_rider_info
```

```python
clean_rider_info =[]

for i in cleaner_rider_info:
    clean_rider_info.append(re.sub("\n", "", i))

print(clean_rider_info)
```

```python
rider_name_age = []

for i in clean_rider_info:
    rider_name_age.append(re.sub(' years old', '', i))

rider_name_age
```

#### Turn the list into a dataframe
The even element is the name, the uneven element is the age. Put them into seperate lists, which then become the columns of the dataframe.

```python
name=[]
age=[]

for index, element in enumerate(rider_name_age):
    if index % 2 == 0:
        name.append(element)
    else:
        age.append(element)

print(name)
print(age)
```

```python
df_rider_info = pd.DataFrame()
df_rider_info['Name']  = name
df_rider_info['Age'] = age

df_rider_info
```

#### Convert dataframe to CSV-file

The page will go offline once the race finishes (~11 may 2021) so it's important to save the data locally.

```python
df_rider_info.to_csv('riders2021_name_age.csv')
```
